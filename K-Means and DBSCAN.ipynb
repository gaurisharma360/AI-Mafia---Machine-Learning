# K-means fails when datasets are large

```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
```

```python
X,y=make_moons(n_samples=200,noise=0.15)
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
plt.show()
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=2)
km.fit(X)
```

```python
centers=km.cluster_centers_
label=km.labels_
```

```python
plt.scatter(X[:,0], X[:,1], c=label)
plt.scatter(centers[:,0], centers[:, 1], color='red')
plt.show()
```

# Implementing DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
#The parameter eps defines the radius of neighborhood around a point x. It's called called the ϵ-neighborhood of x. 
#The parameter MinPts is the minimum number of neighbors within “eps” radius.
dbs=DBSCAN(eps=0.22, min_samples=5)
```

```python
dbs.fit(X)
```

```python
ybs=dbs.fit_predict(X)
```

```python
plt.scatter(X[:,0], X[:, 1], c=ybs)
plt.show()
```
