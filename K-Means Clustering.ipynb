# Implementation of K-means 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
```

```python
X,y=make_blobs(n_samples=1000,n_features=2,centers=7,random_state=10)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
#Initial data preperation
k=7
color=["red", "green", "blue", "yellow", "gray","pink","orange"]
clusters={}
for i in range(k):
    #initialise centres randomly
    center = 10*(2*np.random.random((X.shape[1],))-1)
    points = []
    
    cluster = {
        'center':center,
        'points':points, 
        'color': color[i]
    }
    clusters[i] = cluster
```

```python
def distance(v1,v2):
    return np.sqrt(np.sum((v1-v2)**2))
```

```python
# step 2 
# Assign every data point to one of its clusters 
# E-step 
def assignPointToClusters(clusters):
    
    for i in range(X.shape[0]):
        dist = []
        curr_x = X[i]
        
        for kx in range(k):
            d = distance(curr_x, clusters[kx]['center'])
            dist.append(d)
        
        # minimum distance 
        current_cluster = np.argmin(dist)
        clusters[current_cluster]['points'].append(curr_x)
```

```python
# STEP 3 
# UPDATE CLUSTER CENTER ACCORDING TO MEAN VALUE OF DATA POINTS
# M-Step
def updateClusterCenter(clusters):
    for kx in range(k):
        pts = np.array(clusters[kx]['points'])
        
        if pts.shape[0]>0 :
            mu = pts.mean(axis=0)
            clusters[kx]['center'] = mu
            clusters[kx]['points'] = [] # clear the list
```

```python
def plotClusters(clusters):
    for kx in range(k):
        
        pts = np.array(clusters[kx]['points'])
        
        #plot the points 
        try:
            plt.scatter(pts[:,0], pts[:,1], color=clusters[kx]['color'])
        except:
            pass
        
        # plot the cluster center 
        uk = clusters[kx]['center']
        plt.scatter(uk[0], uk[1], color='black', marker="*")
        
```

```python
assignPointToClusters(clusters)
plotClusters(clusters)
```

```python
# repeat these three functions few times 
count = 1
while True:
    assignPointToClusters(clusters)
    updateClusterCenter(clusters)
    count+=1
    if count>100:
        break
```
